```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')

%matplotlib inline
plt.style.use('ggplot')
```

```python
df = pd.read_csv('weatherAUS.csv')
```

### Data Exploration and Analysis
- Data Exploration
- Visualizations

```python
df.isna().mean().round(4) * 100 #finding null values 
```

- Features Sunshine, Evaporation, Cloud3pm and Cloud9am contains 48%, 43%, 40% and 37% null values respectively.

```python
#Feature Exapantion

df['year'] = [i[:4] for i in df['Date']] #adding year column
df['month'] = [i[5:7] for i in df['Date']] #adding month column
df['day'] = [i[8:10] for i in df['Date']] #adding day column
```

```python
df.hist(figsize=(20,15))
```

```python
df['year'] = df['year'].astype(int)
df['month'] = df['month'].astype(int)
df['day'] = df['day'].astype(int)
```

```python
#Lets look at the distribution of our target variable
sns.countplot(df['RainTomorrow'])
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(df.corr(), annot=True)
```

## Data Preprocessing
    - Categorical Values
    - Missing Values 
    - Feature Scaling
    - Feature Selection
    - Over Sampling
    - Under Sampling

```python
rain_tomorrow_map = {'No': 0, 'Yes': 1}
df['RainTomorrow'] = df['RainTomorrow'].map(rain_tomorrow_map)
```

```python
corr = df.corr()
corr['RainTomorrow']
```

- The features MaxTemp, Temp3pm, Temp9am, Pressure3pm, Pressure9am, Sunshine, Evaporation are negatively correlated with target we variable we can drop these features.

```python
df['RainToday'].value_counts()
df['RainToday'] = df['RainToday'].map(rain_tomorrow_map)
```

```python
#mappings for categorical features WindGistDir, WindDir9am, WindDir3pm
cat_feature_map = {
    'N': 1, 'SE' : 2, 'E': 3, 'SSE': 4, 'NW': 5, 'S': 6,
    'W': 7, 'SW' : 8, 'NNE': 9, 'NNW': 10, 'ENE': 11, 'ESE': 12,
    'NE': 13, 'SSW': 14, 'WNW': 15, 'WSW': 16
}

#Apply mapping to these four columns
for i in ['WindGustDir', 'WindDir9am', 'WindDir3pm']:
    df[i] = df[i].map(cat_feature_map)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Location'] = le.fit_transform(df['Location'])
```

- Feature WindGustDir, WindDir9am, WindDir3pm are categorical features.


## Missing Values
- We're going to use dataframe mean to impute missing values in Dataset.

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(), inplace=True) #Fill null values by mean value.
```

```python
df.info()
```

```python
#We will drop the features that are negatively correlated
df = df.drop(['MaxTemp', 'Temp3pm', 'Temp9am', 'Pressure3pm', 
         'Pressure9am', 'Sunshine', 'Evaporation', 'Date'], axis=1)
```

## Feature Scaling
- We're going to use MinMaxScaler class provided by sklearn to scaling our values between 0 to 1.
- For documentation you can check here --> https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html?highlight=minmaxscaler#sklearn.preprocessing.MinMaxScaler

```python
#Feature scaling
df2 = df.copy()
columns = ['Location', 'MinTemp', 'Rainfall', 'WindGustDir', 'WindGustSpeed',
       'WindDir9am', 'WindDir3pm', 'WindSpeed9am', 'WindSpeed3pm',
       'Humidity9am', 'Humidity3pm', 'Cloud9am', 'Cloud3pm', 'RainToday',
       'RISK_MM', 'RainTomorrow', 'year', 'month', 'day']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_df = scaler.fit_transform(df2)
scaled = pd.DataFrame(scaled_df, columns=columns)
scaled.head(10)
```

## Feature Selection
- For feature selection we are going to use SelectKBest class provided by sklearn.
- SelectKBest class scores the features using a function.
- Here is documentation take look at it --> https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.SelectKBest.html

```python
from sklearn.feature_selection import SelectKBest, chi2

X = scaled.drop('RainTomorrow', axis=1)
y = scaled['RainTomorrow']

bestfeatures = SelectKBest(score_func=chi2, k=15)
fit = bestfeatures.fit(X, y)
scores = pd.DataFrame(fit.scores_)
columns = pd.DataFrame(X.columns)
feature_scores = pd.concat([scores, columns], axis=1)
feature_scores.columns = ['Score', 'Features']
feature_scores.nlargest(10, 'Score')
```

```python
#selecting top 10 features
X = X[feature_scores[['Score','Features']].sort_values(by= 'Score',ascending=False)[:10]['Features'].tolist()]
```

```python
from sklearn.model_selection import train_test_split
```

- Splitting dataset into train and test 80% for training and remaining 20% for test.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state = 42)

print("Training Data: ", X_train.shape, y_train.shape)
print("Testing Data: ",X_test.shape, y_test.shape)
```

- As we have observe in the EDA step that our dataset is imbalance. 
- There are different techniques to handle the imbalance dataset, such as oversampling and undersampling.
- We are going to try both of these method and analyse the result.

## Oversampling
- We are going to use SMOTE algorithm, it stands for Synthetic Minority Oversampling Technique.
- Basically, its statistical technique for increasing the numbers of minority cases in you Dataset.
- For more information you can check documentation -->  https://imbalanced-learn.readthedocs.io/en/stable/generated/imblearn.over_sampling.SMOTE.html

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 42)
X_train_over, y_train_over = sm.fit_sample(X_train, y_train)
```

```python
print("Shape of training data before oversampling:  ", X_train.shape, y_train.shape)
print("Shape of training data after oversampling:  ",X_train_over.shape, y_train_over.shape)
```

## Undersampling
- We are going to use RandomUnderSampler algorithm for undersampling.

```python
#performing undersampling
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state = 42)
X_train_under, y_train_under = rus.fit_sample(X, y)
```

```python
print("Shape of training data before undersampling:  ", X_train.shape, y_train.shape)
print("Shape of training data after undersampling:  ",X_train_under.shape, y_train_under.shape)
```

# Modeling
- We're going to use LogisticRegression, KNeighborsClassifier, DecisionTreeClassifier algorithms for modeling.
- LogisticRegression --> Linear Classifier
- KNeighborsClassifier --> Distance Based
- DecisionTreeClassifier --> Tree-Based

```python
#import all models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
#All metrics using which we are going to test our classifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score
from mlxtend.plotting import plot_confusion_matrix
```

#### LogisticRegression

```python
log_reg = LogisticRegression()
```

```python
#LogisticRegression on Original Data

log_reg.fit(X_train, y_train)
log_norm_pred = log_reg.predict(X_test)
log_norm_score = roc_auc_score(y_test, log_norm_pred)
```

```python
print(classification_report(y_test, log_norm_pred))
print("ROC_AUC_SCORE: ", log_norm_score)
```

```python
plot_confusion_matrix(confusion_matrix(y_test, log_norm_pred))
```

```python
#LogisticRegression with Oversampled Data

log_reg.fit(X_train_over, y_train_over)
log_over_pred = log_reg.predict(X_test)
log_over_score = roc_auc_score(y_test, log_over_pred)
```

```python
print(classification_report(y_test, log_over_pred))
print('ROC_AUC_SCORE: ', roc_auc_score(y_test, log_over_pred))
```

```python
plot_confusion_matrix(confusion_matrix(y_test, log_over_pred))
```

```python
#LogisticRegression with Undersampled Data

log_reg.fit(X_train_under, y_train_under)
log_under_pred = log_reg.predict(X_test)
log_under_score = roc_auc_score(y_test, log_under_pred)
```

```python
print(classification_report(y_test, log_under_pred))
print('ROC_AUC_SCORE: ', roc_auc_score(y_test, log_under_pred))
```

```python
plot_confusion_matrix(confusion_matrix(y_test, log_under_pred))
```

#### LogisticRegression report

- Original Data --> AUC score: 80%
- Oversampled Data --> AUC Score: 89%
- Undersampled Data --> AUC Score: 85%

```python
#KNeighborsClassifier 
knn = KNeighborsClassifier(n_jobs=-1)
param = {'n_neighbors': [5, 10, 15, 20, 25, 30]}
gsc = GridSearchCV(knn, param_grid=param, cv=3)
gsc.fit(X_train, y_train)
print(gsc.best_params_)
```

```python
knn_final = KNeighborsClassifier(**gsc.best_params_, n_jobs=-1)
knn_final.fit(X_train, y_train)
knn_norm_pred = knn_final.predict(X_test)
knn_norm_score = roc_auc_score(y_test, knn_norm_pred)
```

```python
print(classification_report(y_test, knn_norm_pred))
print("AUC: ", knn_norm_score)
```

```python
plot_confusion_matrix(confusion_matrix(y_test, knn_norm_pred))
```

```python
#kNN on Oversampled Data
knn_final.fit(X_train_over, y_train_over)
knn_over_pred = knn_final.predict(X_test)
knn_over_score = roc_auc_score(y_test, knn_over_pred)
```

```python
print(classification_report(y_test, knn_over_pred))
print("AUC: ", knn_over_score)
```

```python
plot_confusion_matrix(confusion_matrix(y_test, knn_over_pred))
```

```python
#kNN on undersampled Data
knn_final.fit(X_train_under, y_train_under)
knn_under_pred = knn_final.predict(X_test)
knn_under_score = roc_auc_score(y_test, knn_under_pred)
```

```python
print(classification_report(y_test, knn_under_pred))
print("AUC:  {0:.2f}".format(knn_under_score))
```

```python
plot_confusion_matrix(confusion_matrix(y_test, knn_under_pred))
```

#### KNN Report#### 

- Original Data --> AUC score: 70%
- Oversampled Data --> AUC Score: 76%
- Undersampled Data --> AUC Score: 78% 

```python
%%time
#DecisionTreeClassifier
dtc = DecisionTreeClassifier()
param = {
    'max_depth': [2,4,8,10],
    'max_features': [4, 6, 10]
}

grid = GridSearchCV(dtc, param_grid=param, n_jobs=-1, cv=3)
grid.fit(X_train, y_train)
print(grid.best_params_)
```

```python
print(classification_report(y_test, dc_over_pred))
print("ROC_AUC_SCORE: {0:.2f}".format(dc_over_score))
```

```python
plot_confusion_matrix(confusion_matrix(y_test, dc_over_pred))
```

```python
# DC on undersampled data
dc.fit(X_train_under, y_train_under)
dc_under_pred = dc.predict(X_test)
dc_under_score = roc_auc_score(y_test, dc_under_pred)
```
